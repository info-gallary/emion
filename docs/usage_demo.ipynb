# ⚛️ EmION Usage Examples
This notebook demonstrates how to interact with the EmION framework programmatically.

## 1. Core BPv7 Orchestration
First, we initialize the ION core and register nodes.

```python
import emion
from emion.core import network

# Create a 3-node topology
nodes = [1, 2, 3]
for nid in nodes:
    network.register_node(nid)

# Start the ION engine
network.start_core()
print("ION Core Initialized with nodes:", nodes)
```

## 2. Dispatching Bundles
Sending a mission-critical bundle from Node 1 to Node 3.

```python
payload = "MISSION_B_V7_DATA"
result = network.send_bundle(src=1, dst=3, payload=payload)
print(f"Bundle Dispatched! ID: {result.get('bundle_id')}")
```

## 3. Anomaly Module Integration
Attaching an external security module to monitor traffic.

```python
module_url = "http://localhost:8421"
network.attach_plugin(module_url, target_nodes="all")
print(f"Module {module_url} is now monitoring the network.")
```
